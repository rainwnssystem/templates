```python
import mlflow

MLFLOW_ARN = "arn:aws:sagemaker:us-east-1:200148130345:mlflow-app/app-KXKWC5KLMVDB"
EXPERIMENT_NAME = "iris-xgb-classifiers"

mlflow.set_tracking_uri(MLFLOW_ARN)
mlflow.set_experiment(EXPERIMENT_NAME)
print("Tracking URI: ", mlflow.get_tracking_uri())
```

```python
%%bash

curl -L -o ./iris.zip https://www.kaggle.com/api/v1/datasets/download/uciml/iris
unzip -q iris.zip
```

```python
import sagemaker

sess = sagemaker.Session()
DEFAULT_BUCKET = sess.default_bucket()    # sagemaker-us-east-1-<account-id>
print("Default bucket:", DEFAULT_BUCKET)

iris_s3_uri = sess.upload_data(
    path="./iris.zip",
    bucket=DEFAULT_BUCKET
)
print("Training input channel:", iris_s3_uri)
```

```python
from sagemaker.xgboost import XGBoost

role = sagemaker.get_execution_role()
print("Execution role:", role)

hp_grid = [
    {"n_estimators":  50, "max_depth": 3, "learning_rate": 0.10},
    {"n_estimators": 100, "max_depth": 5, "learning_rate": 0.10},
    {"n_estimators": 200, "max_depth": 7, "learning_rate": 0.05},
]

for i, hp in enumerate(hp_grid):
    estimator = XGBoost(
        entry_point="train.py",
        source_dir="src",
        role=role,
        instance_type="ml.m5.large",
        instance_count=1,
        framework_version="3.0-5",
        base_job_name=f"mlflow-iris-{i+1}",
        hyperparameters={
            **hp,
            "tracking_uri": MLFLOW_ARN,
            "experiment_name": EXPERIMENT_NAME,
        },
    )
    estimator.fit(inputs={"train": iris_s3_uri}, wait=True)
    print(f"Job {i+1} completed.")
```
